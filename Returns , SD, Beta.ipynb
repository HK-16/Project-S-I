```python
import pandas as pd
import numpy as np
df = pd.read_excel('MASTER FILE WITH NIFTY.xlsx')
df.set_index('Date',inplace=True)
display(df)
returns = df.pct_change()
cov = returns.cov()
var = returns['NIFTY'].var()
print(var)
```

```python
beta = (round(cov.loc['NIFTY']/var,2))
beta = beta.to_frame(name="Beta")
# beta.to_excel('Beta.xlsx')
display(beta)
```

```python
print('Returns Annually')
annualreturns = (round(returns.mean()*252*100,2))
annualreturns = annualreturns.to_frame(name="Annual Returns (%)")
# annualreturns.to_excel('Average Annual Returns.xlsx')
display(annualreturns)
```

```python
std = round(returns.std()*np.sqrt(252)*100,2)
std = std.to_frame(name='Annual Standard Deviation (%)')
# std.to_excel('Annual Standard Deviation.xlsx')
display(std)
```
